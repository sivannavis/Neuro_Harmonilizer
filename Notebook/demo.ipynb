## Training our own tension harmonilizer neural networks

### Prepare datasets for training inputs

```python
# get datasets
database = "/Users/sivanding/database/jazznet/chords"
metadata = "/Users/sivanding/database/jazznet/metadata/tiny.csv"
train_data, _ = prepare_data(database, metadata, 'train')
val_data, _ = prepare_data(database, metadata, 'validation')
test_data, _ = prepare_data(database, metadata, 'test')
test_ft, test_or, test_ten = test_data
```

```python
print(f'#test data{test_ft.shape[0]}\n')
```

### Finetuning pretrained crema + a naive classifier to map from embedding to harmony colors

Classifier architecture is referred from OPENL3 : https://github.com/marl/openl3

```python
# get model
model = tension_model()
model.summary()
```

```python
# train and validation - takes ~ 25min
train(model, train_data, val_data, output_dir='train_03', epochs=100, batch_size=10)
```

```python
# testing
scores = model.evaluate(test_ft, [test_or, test_ten], batch_size=10)
print(scores)
```

### How about fixing the pretrained crema and only train the classifier from scratch?

```python
# get model
model2 = fixed_tension_model()
model2.summary()
```

```python
# train and validation - takes ~ 25min
train(model2, train_data, val_data, output_dir='train_02', epochs=100, batch_size=10)
```

```python
# testing
scores2 = model2.evaluate(test_ft, [test_or, test_ten], batch_size=10)
print(scores2)
```

## Training Process Diagnostics

```python
import matplotlib.pyplot as plt
%matplotlib nbagg
import pandas as pd
```

```python
output_dir = "train_01"
with open(f'./models/history_{output_dir}', 'rb') as file_pi:
    history = pickle.load(file_pi)
history = pd.DataFrame.from_dict(history)
history
```

```python
# two losses during training and val
plt.figure()
plt.plot(history['loss'], label='Training loss')
plt.plot(history['val_loss'], label='Validation loss')
plt.legend(loc='best')
plt.tight_layout()
# two metrics during training and val
plt.figure()
plt.plot(history['time_distributed_1_categorical_accuracy'], label='Training accuracy')
plt.plot(history['val_time_distributed_1_categorical_accuracy'], label='Validation accuracy')
plt.legend(loc='best')
plt.tight_layout()

plt.figure()
plt.plot(history['time_distributed_mean_squared_error'], label='Training mse')
plt.plot(history['val_time_distributed_mean_squared_error'], label='Validation mse')
plt.legend(loc='best')
plt.tight_layout()
```
