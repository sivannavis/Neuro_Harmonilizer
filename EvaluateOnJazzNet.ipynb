```python
import os
import autochord
# Folder containing your audio files
folder_path = "/content/drive/MyDrive/Untitled_folder/progressions-small"
```

```python
def process_audio_files(folder_path):
    results = {}

    # List all files in the directory
    for filename in os.listdir(folder_path):
        if filename.endswith('.wav'):  # Check if the file is a WAV audio file
            file_path = os.path.join(folder_path, filename)

            try:
                # Process the file with autochord and store the output
                chords = autochord.recognize(file_path)
                results[file_path] = chords  # Store the list of tuples for each file
                print(f"Processed: {file_path}")
            except Exception as e:
                print(f"Error processing {file_path}: {e}")
                results[file_path] = None  # Add None for files that encountered errors

    return results
```

```python
folder_path = "/content/drive/MyDrive/Untitled_folder/progressions-small"
```

```python
processed_results = process_audio_files(folder_path)
```

```python
print(processed_results)
```

```python
# Convert the Dictionary to CSV Format
import csv

def convert_to_csv_format(processed_results):
    csv_data = []
    for file_path, chords in processed_results.items():
        if chords is not None:
            for start, end, chord_name in chords:
                csv_data.append([file_path, start, end, chord_name])
        else:
            csv_data.append([file_path, None, None, None])  # Handle files with errors
    return csv_data
```

```python
# Write to a CSV File
csv_data = convert_to_csv_format(processed_results)

with open('chord_results.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['File Path', 'Chord Start', 'Chord End', 'Chord Name'])  # Header
    writer.writerows(csv_data)
```
